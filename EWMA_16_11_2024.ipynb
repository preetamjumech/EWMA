```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DailyDelhiClimateTrain.csv')
df.head()
```

```python
df = df[["date", "meantemp"]]
df.head()
```

```python
plt.scatter(df["date"], df["meantemp"], color = 'red')
```

```python
x1 = df["meantemp"].ewm(alpha=0.9).mean() # alpha = 1- beta
x1.head()
```

```python
df["ewma"] = x1
df.head()
```

```python
plt.scatter(df["date"], df["meantemp"], color = 'red')
plt.plot(df["date"], df["ewma"], color = 'black', label = 'alpha = 0.9')
plt.legend()
plt.show()
```
